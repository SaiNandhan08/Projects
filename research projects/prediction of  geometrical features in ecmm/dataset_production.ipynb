```python
import os
import cv2
import numpy as np
import pandas as pd
import mahotas as mt
import h5py
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def create_dataset():
    names = ['Input','Target']
    df = pd.DataFrame([], columns=names)
    total = 0
    while total < 200:
        total += 1 
        
        
        import random
        num1 = 0
        num1 = random.randint(0,9)
        num2 = random.randint(0,1000)
        num3 = num2/1000
        num1 = num1 + num3
        
        
        reg_coeff = 5
        reg_intercept = 40
        target = 0
        target = num1/reg_coeff
        target = target + reg_intercept
        
        
        Input = num1
        
        Target = target       
        vector = [Input,Target]
        df_temp = pd.DataFrame([vector],columns=names)
        df = df.append(df_temp)
        
    return df
```

```python
dataset = create_dataset()
```
