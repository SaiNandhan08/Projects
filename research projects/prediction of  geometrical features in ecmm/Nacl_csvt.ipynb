```python
import os
import cv2
import numpy as np
import pandas as pd
import mahotas as mt
import h5py
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def create_dataset():
    names = ['Voltage', 'Feed Rate ', 'Duty cycle','MRR','Circularity', 'Conicity', 'Overcut']
    df = pd.DataFrame([], columns=names)
    total = 0
    while total < 130:
        total += 1 
        
        
        import random
        num1 = 0
        num1 = random.randint(6,8)
        
        num2 = random.randint(5,7)
        num3 = num2/10
        
        num4 = random.randint(6,8)
        num5 = num4/10
        
        num6 = random.randint(220,410)
        num7 = num6/1000000
        
        
        rc1 = 0.000055
        rc2 = -0.022103
        rc3 = 0.050021
        rc4 = -40.715775
        ri1= -0.0010519181471904896
        
        rc5 = -0.016906
        rc6 = -0.963664
        rc7 = 12.190442
        rc8 = -12919.204943
        ri2 = -2.6591530989351577
        
        rc9 = 0.031481
        rc10 = -0.049851
        rc11 = -0.438901
        rc12 = 1190.777752
        ri3 = 0.48930651005674153
        
        Cir = 0
        Cir = num1*rc1
        Cir = Cir + num3*rc2
        Cir = Cir + num5*rc3
        Cir = Cir + num7*rc4
        Cir = Cir + ri1
        Cir = round(Cir, 3)
        
        
        Con = 0
        Con = num1*rc5
        Con = Con + num3*rc6
        Con = Con + num5*rc7
        Con = Con + num7*rc8
        Con = Con + ri2
        Con = round(Con, 3)
        
        
        Ove = 0
        Ove = num1*rc9
        Ove = Ove + num3*rc10
        Ove = Ove + num5*rc11
        Ove = Ove + num7*rc12
        Ove = Ove + ri3
        Ove = round(Ove, 4)
        
        
        I1 = num1
        I2 = num3
        I3 = num5
        I4 = num7
        
        Target1 = Cir
        Target2 = Con
        Target3 = Ove
        
        
               
        vector = [I1,I2,I3,I4,Target1,Target2,Target3]
        
        df_temp = pd.DataFrame([vector],columns=names)
        df = df.append(df_temp)
        
    return df
```

```python
dataset = create_dataset()
```
